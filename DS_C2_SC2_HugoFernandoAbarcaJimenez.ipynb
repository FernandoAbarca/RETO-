**a. Extraer la información del archivo.**

```python
import pandas as pd
country_vaccinations=pd.read_csv('/content/drive/MyDrive/RETO COVID/country_vaccinations.csv')
country_vaccinations_manufacturer=pd.read_csv('/content/drive/MyDrive/RETO COVID/country_vaccinations_by_manufacturer.csv')
```

**b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64. **

```python
country_vaccinations['date']=pd.to_datetime(country_vaccinations['date'])
country_vaccinations.info()
```

```python
country_vaccinations_manufacturer['date']=pd.to_datetime(country_vaccinations_manufacturer['date'])
country_vaccinations_manufacturer.info()
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines). 

```python
country_vaccinations_manufacturer[['vaccine','total_vaccinations']].groupby('vaccine').sum()
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo. 

```python
print('Total de vacunas aplicadas en el mundo:',country_vaccinations['people_vaccinated'].sum())
```

e. Calcular el promedio de vacunas aplicadas por país. 

```python
country_vaccinations[['country','people_vaccinated']].groupby('country').mean()
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo. 

```python
total=country_vaccinations[country_vaccinations['date'].isin(['2021-01-29'])]
total.sample(3)
```

```python
print('Personas vacunadas el 29-01-2021 en todo el mundo:',total['people_vaccinated'].sum())
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw. 

```python
conDiferencias=country_vaccinations.copy()
```

```python
conDiferencias['diferencias']=conDiferencias['daily_vaccinations']-conDiferencias['daily_vaccinations_raw']
```

```python
conDiferencias.sample()
```

h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
print('El periodo de tiempo entre registro más reciente y registro más antiguo es:',conDiferencias['date'].max()-conDiferencias['date'].min())
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 

```python
conCantidad=country_vaccinations.copy()
```

```python
conCantidad['canVac']=conCantidad['vaccines'].str.split('/').str.len()
```

```python
conCantidad.sample()
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20=country_vaccinations.query('date<="2020-12-20"').copy()
antes20.sample(3)
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 

```python
pfizer=country_vaccinations[country_vaccinations['vaccines'].str.contains('pfizer',case=False)].copy()
pfizer.shape
```

```python
pfizer.sample()
```

```python
country_vaccinations.shape
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
with pd.ExcelWriter('/content/drive/MyDrive/RETO COVID/resultadosReto.xlsx') as base:
  conDiferencias.to_excel(base, sheet_name='conDiferencia')
  conCantidad.to_excel(base, sheet_name='conCantidad')
  pfizer.to_excel(base, sheet_name='pfizer')
  antes20.to_excel(base, sheet_name='antes20')



```
